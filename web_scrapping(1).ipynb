```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
}

webpage = requests.get("https://www.ambitionbox.com/list-of-companies?page=1",headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
# fetching data whole data of the company 
company = soup.find_all("div",class_= "companyCardWrapper")
len(company)
```

```python
name = []
rating = []
reviews = []
salaries = []
interviews = []
jobs = []
benefits = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find_all('span',class_='companyCardWrapper__companyRatingValue')[0].text.strip())
    reviews.append(i.find_all('span',class_="companyCardWrapper__ActionCount")[0].text.strip())
    salaries.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[1].text.strip())
    interviews.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[2].text.strip())
    jobs.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[3].text.strip())
    benefits.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[4].text.strip())
    # salaries.append(i.find_all("span",class_="companyCardWrapper__ActionCount")[5].text.strip())

d= {"name" : name , "ratings":rating , "reviews" : reviews ,"salaries" :salaries , "interviews":interviews,"jobs":jobs , "benefits" : benefits}
df = pd.DataFrame(d)
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final = pd.DataFrame()

for j in range(1, 6):
    url = f"https://www.ambitionbox.com/list-of-companies?page={j}"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all('div', class_='company-content-wrapper')

    name = []
    rating = []
    reviews = []
    salaries = []
    interviews = []
    jobs = []
    benefits = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('span', class_='company-rating').text.strip())
        reviews.append(i.find('span', class_='reviews').text.strip())
        salaries.append(i.find('span', class_='salaries').text.strip())
        interviews.append(i.find('span', class_='interviews').text.strip())
        jobs.append(i.find('span', class_='jobs').text.strip())
        benefits.append(i.find('span', class_='benefits').text.strip())

    data = {
        "Name": name,
        "Rating": rating,
        "Reviews": reviews,
        "Salaries": salaries,
        "Interviews": interviews,
        "Jobs": jobs,
        "Benefits": benefits
    }

    df = pd.DataFrame(data)
    final = pd.concat([final, df], ignore_index=True)

print(final.to_csv("company_data.csv", index=False))
```
